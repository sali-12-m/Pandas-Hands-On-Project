  - Sales

- The dataset is suitable for:
  - Sales analysis
  - Customer behavior analysis
  - Profitability evaluation

```python
df.isnull().sum()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

#  Task 2: Data Cleaning – Insights

## What we observed:
- Very few or no critical missing values.
- Date columns were stored as text initially.

## Why cleaning matters:
- Converting `Order Date` and `Ship Date` to datetime allows:
  - Time-based analysis (monthly, yearly trends)
  - Accurate grouping and filtering

## Key Insight:
- Clean data ensures **accurate analysis and avoids errors**.
- Without correct data types, time-series analysis would not be possible.


```python
df["Month"] = df["Order Date"].dt.month
df["Year"] = df["Order Date"].dt.year
```

```python
df["Profit Margin"] = df["Profit"] / df["Sales"]
```

```python
df[["Sales", "Profit", "Profit Margin", "Month", "Year"]].head()
```


#  Task 3: Feature Engineering – Insights

## What we created:
- `Month` and `Year` from Order Date
- `Profit Margin = Profit / Sales`

## Why this is useful:

### Month & Year:
- Helps identify **seasonality**
- Makes trend analysis easier

### Profit Margin:
- Shows efficiency, not just revenue
- Example:
  - High sales + low profit = bad business decision

## Key Insight:
- New features reveal **hidden patterns** not visible in raw data.

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Region")["Profit"].sum()
```

```python
df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(5)
```

```python
monthly_sales = df.groupby(pd.Grouper(key="Order Date", freq="M"))["Sales"].sum()
monthly_sales
```
